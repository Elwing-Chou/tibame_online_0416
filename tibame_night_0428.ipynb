```python
import urllib.request as req

url = "https://github.com/Elwing-Chou/tibame_online_0416/raw/refs/heads/main/tabelog%20(3).csv"
req.urlretrieve(url, "tabelog.csv")
```

```python
import pandas as pd
df = pd.read_csv("tabelog.csv", encoding="utf-8", index_col=0)
df
```

```python
# 上次的!!!
def func(score):
    return round(score * 2, 1)
df["評分"].apply(func)
```

```python
# 我要變成七個行(禮拜一open: Y/N)
# pandas: DataFrame(表格)(2D)  Series(行/列)(1D)
# 如果你想把值擴充(多個欄位): return Series
def func(s):
    open_days = {
        "星期一":"Y",
        "星期二":"Y",
        "星期三":"Y",
        "星期四":"Y",
        "星期五":"Y",
        "星期六":"Y",
        "星期日":"Y",
    }
    for spl in s.split("、"):
        if spl in open_days:
            open_days[spl] = "N"
    return pd.Series(open_days)

# 方法1: df[["星期一","星期二","星期三"...]] = 轉換
# pandas的功能大部分都有一個參數 axis 只有兩種值 0/1 剛好代表兩種方向
convert = df["休息日"].apply(func)
df = pd.concat([df, convert], axis=1)
df
```

```python
# 過濾: 帶入跟你資料筆數一樣多的True/False
query_day = "星期三"
fil = df[query_day] == "Y"
df[fil]
```
