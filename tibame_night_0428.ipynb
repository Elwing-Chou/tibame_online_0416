````python
import urllib.request as req

url = "https://github.com/Elwing-Chou/tibame_online_0416/raw/refs/heads/main/tabelog%20(3).csv"
req.urlretrieve(url, "tabelog.csv")
````

````python
import pandas as pd
df = pd.read_csv("tabelog.csv", encoding="utf-8", index_col=0)
df
````

````python
# 上次的!!!
def func(score):
    return round(score * 2, 1)
df["評分"].apply(func)
````

````python
# 我要變成七個行(禮拜一open: Y/N)
# pandas: DataFrame(表格)(2D)  Series(行/列)(1D)
# 如果你想把值擴充(多個欄位): return Series
def func(s):
    open_days = {
        "星期一":"Y",
        "星期二":"Y",
        "星期三":"Y",
        "星期四":"Y",
        "星期五":"Y",
        "星期六":"Y",
        "星期日":"Y",
    }
    for spl in s.split("、"):
        if spl in open_days:
            open_days[spl] = "N"
    return pd.Series(open_days)

# 方法1: df[["星期一","星期二","星期三"...]] = 轉換
# pandas的功能大部分都有一個參數 axis 只有兩種值 0/1 剛好代表兩種方向
convert = df["休息日"].apply(func)
df = pd.concat([df, convert], axis=1)
df
````

````python
# 過濾: 帶入跟你資料筆數一樣多的True/False
query_day = "星期三"
fil = df[query_day] == "Y"
df[fil]
````

# 下載圖片



```
HTTP代碼:
1. 2開頭: 成功
2. 3開頭: 轉址(輸入舊網址, 帶你來到新的)
3. 4開頭: 錯誤
    a. 404: Not Found(檢查網址)
    b. 403: Forbidden
        - 禁止ip(很多次以後)
        - 你跟瀏覽器不一樣(!!!!)

一個網址(額外訊息: request headers)
一個回覆(額外訊息: response headers)
```



````python
import urllib.request as req
import bs4 as bs

url = "https://www.ptt.cc/bbs/Beauty/M.1745582404.A.011.html"
r = req.Request(url)
r.add_header("User-Agent", "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0")
resp = req.urlopen(r)
html = bs.BeautifulSoup(resp.read())
print(html)
````
